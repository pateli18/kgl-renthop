# Feature Selection

```python
features = set()

def extract_features(feature_list):
    global features
    if len(feature_list) > 0:
        features = set(feature_list) | features
    
train_df['features'].apply(lambda features: extract_features(features))
features = {feature for feature in features if '*' not in feature}
```
